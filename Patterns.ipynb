```python
rows=5
for i in range(rows):
    for j in range(rows):
        print("*",end=" ")
    print()
```

```python
rows=5
for i in range(rows):
    for j in range(rows):
        if i==0 or i==rows-1 or j==0 or j==rows-1:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
rows=5
for i in range(1,rows+1):
    print((str(i)+" ")*i)
```

```python
rows=5
for i in range(1,rows+1):
    for j in range(1,i+1):
        print(j,end=" ")
    print()
```

```python
rows=5
for i in range(1,rows+1):
    ch=ord('A')
    for j in range(i):
        print(chr(ch),end=" ")
        ch+=1
    print()
```

```python
rows=5
for i in range(rows):
    for j in range(i+1):
        print(chr(65+i),end=" ")
    print()
```
