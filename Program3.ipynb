Program3: Write a program to display the elements of the matrix X to different powers and identity matrix of a given matrix.Also create another matrix Y with same dimensions and display X2+2Y

```python
import numpy as np;
X=np.array([[1,2],
[3,4]])
Y=np.array([[5,6],[7,8]])
print("Matrix X is:\n ",X)
print("Matrix Y is: \n",Y)
a=np.power(X,2)
print("X^2=",a)
result=a+2*Y
print("X^2+2*Y is \n ",result)
```
